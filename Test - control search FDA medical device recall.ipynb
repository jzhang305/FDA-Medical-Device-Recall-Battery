# Function test - choose recall date range and search

```python
import requests
from bs4 import BeautifulSoup

# Base URL of the FDA medical device recalls website
base_url = 'https://www.accessdata.fda.gov/scripts/cdrh/cfdocs/cfRES/res.cfm'  # Replace with the actual URL

# Set the recall date range
recall_date_from = '11/01/2023'  # Replace with your desired start date
recall_date_to = '11/15/2023'  # Replace with your desired end date

# Create a session
session = requests.Session()

# Send a GET request to the website to fetch the initial page
response = session.get(base_url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Extract the form data and update the recall date range
form_data = {}
for input_tag in soup.find_all('input'):
    input_name = input_tag.get('name')
    if input_name == 'postdatefrom':
        input_tag['value'] = recall_date_from
    elif input_name == 'postdateto':
        input_tag['value'] = recall_date_to

    form_data[input_name] = input_tag.get('value', '')

# Submit the form by sending a POST request with the updated data
response = session.post(base_url, data=form_data)

# Parse the search results page
result_soup = BeautifulSoup(response.text, 'html.parser')

# Process and extract the desired information from the search results
# ... (Add your code to extract the relevant data from the result_soup)

search_results = result_soup.find_all('a', style='text-decoration:underline; color: #23238e;')
for result in search_results:
    print(result.text)

# Close the session
session.close()
```

```python
import time
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
def obt_monthly_recall(month,year):
    # Set the path to the EdgeDriver executable
    edge_driver_path = './msedgedriver.exe'

    # Set the desired location to save the CSV file
    csv_save_location = './save data'

    # Create a new instance of the Edge driver
    service = Service(edge_driver_path)
    driver = webdriver.Edge(service=service)

    # Load the website
    driver.get('https://www.accessdata.fda.gov/scripts/cdrh/cfdocs/cfRES/res.cfm')  # Replace with the actual URL

    # Find and input the recall date range
    start_date = str(month)+('/01/')+str(year)
    
    if month in [1,3,5,7,10,12]:
        enddateofmon =31
    elif (month ==2)&(year % 4 == 0):
        enddateofmon = 29
    elif  (month ==2)&(year % 4 != 0):   
        enddateofmon = 28
    else:
        enddateofmon = 30
    
    end_date =str(month)+str('/')+str(enddateofmon)+str('/')+str(year)    
    
    driver.find_element(By.ID, 'postdatefrom').send_keys(start_date)
    driver.find_element(By.ID, 'postdateto').send_keys(end_date)

    # Click the "Search" button
    driver.find_element(By.NAME, 'Search').click()

    # Click the "Export to Excel" button
    export_button = driver.find_element(By.XPATH, "//a[contains(@title, 'Export to Excel')]")
    driver.execute_script("arguments[0].click();", export_button)

    # Wait for 3 seconds
    time.sleep(3)

    # Close the browser
    driver.quit()
```

# Download data from 2010 to 2022

```python
for year in range(2010,2023):
    for month in range(1,13):
        obt_monthly_recall(month,year)
```

```python
import os
import pandas as pd

# Set the path to the folder containing the CSV files
folder_path = './save data'

# Get a list of all CSV file paths in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Initialize an empty list to store dataframes
dataframes = []

# Iterate over each CSV file
for file in csv_files:
    # Construct the full file path
    file_path = os.path.join(folder_path, file)
    
    # Read the CSV file into a dataframe
    df = pd.read_excel(file_path)
    
    # Append the dataframe to the list
    dataframes.append(df)

# Concatenate all dataframes into one
combined_df = pd.concat(dataframes, ignore_index=True)

combined_df
```
